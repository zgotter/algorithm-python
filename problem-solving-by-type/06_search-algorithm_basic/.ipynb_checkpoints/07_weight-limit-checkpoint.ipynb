### 7.1.1 문제

N(2 ≤ N ≤ 10,000)개의 섬으로 이루어진 나라가 있다. 이들 중 몇 개의 섬 사이에는 다리가 설치되어 있어서 차들이 다닐 수 있다.

영식 중공업에서는 두 개의 섬에 공장을 세워 두고 물품을 생산하는 일을 하고 있다. 물품을 생산하다 보면 공장에서 다른 공장으로 생산 중이던 물품을 수송해야 할 일이 생기곤 한다. 그런데 각각의 다리마다 중량제한이 있기 때문에 무턱대고 물품을 옮길 순 없다. 만약 중량제한을 초과하는 양의 물품이 다리를 지나게 되면 다리가 무너지게 된다.

한 번의 이동에서 옮길 수 있는 물품들의 중량의 최댓값을 구하는 프로그램을 작성하시오.

<br>

### 7.1.2 입력

첫째 줄에 N, M(1 ≤ M ≤ 100,000)이 주어진다. 다음 M개의 줄에는 다리에 대한 정보를 나타내는 세 정수 A, B(1 ≤ A, B ≤ N), C(1 ≤ C ≤ 1,000,000,000)가 주어진다. 이는 A번 섬과 B번 섬 사이에 중량제한이 C인 다리가 존재한다는 의미이다. 서로 같은 두 도시 사이에 여러 개의 다리가 있을 수도 있으며, 모든 다리는 양방향이다. 마지막 줄에는 공장이 위치해 있는 섬의 번호를 나타내는 서로 다른 두 정수가 주어진다. 공장이 있는 두 섬을 연결하는 경로는 항상 존재하는 데이터만 입력으로 주어진다.

<br>

### 7.1.3 출력

첫째 줄에 답을 출력한다.

<br>

### 7.1.4 입출력 예제

**예제 입력 1**

```
3 3

1 3
```

3

## 7.3 해설

### 7.3.1 해설 내용

- 다리의 개수 M은 최대 100,000이며, 중량 제한 C는 최대 1,000,000,000이다.
- 이진 탐색을 이용하여 $O \left( M log C \right)$에 문제를 해결할 수 있다.
  - 값이 10억으로 매우 큰 중량 제한 C를 이진 탐색을 통해 탐색해야 시간 복잡도를 $log C$로 줄일 수 있다.
- 한 번의 이동에서 옮길 수 있는 물품들의 중량의 최댓값을 이진 탐색으로 찾는다.  
  
  
- **BFS**를 활용해서 그래프가 주어 졌을 때 특정 노드에서 출발하여 다른 특정 노드로 도착하는 방법을 확인할 수 있다.
- BFS는 보통 다리의 개수(M)만큼 반복 수행된다.
  - $O(M log C) = 100,000 \times log 1,000,000,000 \approx 3,000,000$

- 반복적으로 중량을 설정하며, 이동이 가능한 경우를 찾는다. (시작 노드 : `1`, 도착 노드 : `3`)

- 최소 중량으로 결과값을 초기화한다.

- 최소 중량과 최대 중량의 중간값에 해당되는 5를 중량값으로 설정한다.
- 그런 다음 BFS를 수행하여 1에서 3으로 이동이 가능한 지 확인한다.

- 이동이 가능함을 확인하고 더 큰 중량에서 물품을 이동할 수 있는 지 확인하기 위해 중량값을 증가시킨다.
- 최소 중량값을 `현재 중량값(5)+1` 값으로 바꾼다.

- 최소 중량과 최대 중량의 중간값인 7을 중량값으로 설정한다.
- 그런 다음 BFS를 수행하여 1에서 3으로 이동이 가능한 지 확인한다.

- 그런 다음 BFS를 수행하여 1에서 3으로 이동이 가능한 지 확인한다.

### 7.3.2 해설 코드

````python
from collections import deque

n, m = map(int, input().split())
adj = [[] for _ in range(n+1)]
start = 1000000000 # c
end = 1

for _ in range(m):
    x, y, weight = map(int, input().split())
    adj[x].append((y, weight))
    adj[y].append((x, weight))
    start = min(start, weight)
    end = max(end, weight)

start_node, end_node = map(int, input().split())

result = start

# BFS : 경로가 있는 지 탐색하는 부분
def bfs(c):
    queue = deque([start_node])
    visited = [False] * (n+1)
    visited[start_node] = True
    
    while queue:
        x = queue.popleft()
        for y, weight in adj[x]:
            if not visited[y] and weight >= c:
                visited[y] = True
                queue.append(y)
    
    return visited[end_node]

# 이진 탐색을 수행하는 부분
while (start <= end):
    mid = (start + end) // 2
    if bfs(mid):
        result = mid
        start = mid + 1
    else:
        end = mid - 1
        
print(result)
````
