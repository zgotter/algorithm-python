# 1. DFS와 BFS (복습)

```python
from collections import deque

def dfs(v):
    arr = graph[v]
    visited[v] = True
    result.append(v)
    
    for a in arr:
        if not visited[a]:
            dfs(a)

def bfs(v):
    q = deque()
    q.append(v)
    visited[v] = True
    result.append(v)
    
    while q:
        qval = q.popleft()
        for a in graph[qval]:
            if not visited[a]:
                q.append(a)
                
            
n, m, v = map(int, input().split())
graph = [[] for _ in range(n+1)]

for _ in range(m):
    x, y = map(int, input().split())
    graph[x].append(y)
    graph[y].append(x)
    
for g in graph:
    g.sort()
    
    
visited = [False] * (n+1)
result = []
dfs(v)
print(' '.join(map(str, result)))

visited = [False] * (n+1)
result = []
bfs(v)
print(' '.join(map(str, result)))
```
