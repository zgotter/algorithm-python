# 1. 배열 (Array)

- 데이터를 나열하고, 각 데이터를 **인덱스**에 대응하도록 구성한 데이터 구조
- 파이썬에서는 리스트 타입이 배열 기능을 제공하고 있음

<br>

## 1.1 배열이 필요한 이유

- **같은 종류**의 데이터를 효율적으로 관리하기 위해 사용
- 같은 종류의 데이터를 순차적으로 저장

<br>

## 1.2 배열의 장점

- 빠른 접근 가능

<br>

## 1.3 배열의 단점

- 배열의 크기를 미리 설정해야 한다.
- 가변적인 데이터에 대해 추가 삭제가 쉽지 않다. (중간이 빌 경우 뒤에 데이터가 앞으로 당겨짐)
- 파이썬 리스트의 경우 일반적인 배열보다 많은 기능을 제공한다.

<br>

## 1.4 파이썬에서의 배열 사용

```python
country = "US"
print(country)
```

```python
country = country + "A" # 일반적인 배열은 파이썬의 리스트와 달리 길이를 미리 지정해야 한다.
print(country)
```
