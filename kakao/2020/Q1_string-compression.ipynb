# 1. 문자열 압축

```python
str1 = "aabbaccc" # 7
str2 = "ababcdcdababcdcd" # 9
str3 = "abcabcdede" # 8
str4 = "abcabcabcabcdededededede" # 14
str5 = "xababcdcdababcdcd" # 17
```

```python
def solution(s):
    answer = len(s)
    result = ""
    for cut in range(1, (len(s)//2)+1):
        print('cut : ', cut)
        count = 1
        tempStr = s[:cut]
        for i in range(cut, len(s), cut):
            if s[i:i+cut] == tempStr:
                count += 1
            else:
                if count == 1:
                    count = ""
                result += str(count) + tempStr
                tempStr = s[i:i+cut]
                print('result : ', result)
                count = 1
                
        if count == 1:
            count = ""
        result += str(count) + tempStr
        print('result : ', result)
        if answer > len(result):
            answer = len(result)
        result = ""
        
    return answer
```

```python
print(solution(str1)) # str1 = "aabbaccc"
```

```python
str2 = "ababcdcdababcdcd" # 9
print(solution(str2))
```

```python
str3 = "abcabcdede" # 8
print(solution(str3))
```

```python
str4 = "abcabcabcabcdededededede" # 14
print(solution(str4))
```

```python
str5 = "xababcdcdababcdcd" # 17
print(solution(str5))
```

```python
def solution2(s):
    length = []
    result = ""
    
    if len(s) == 1:
        return 1
    
    for cut in range(1, (len(s)//2)+1):
        count = 1
        tempStr = s[:cut]
        
        for i in range(1, len(s), cut):
            if s[i:i+cut] == tempStr:
                count += 1
            else:
                if count == 1:
                    count = ""
                result += str(count) + tempStr
                tempStr = s[i:i+cut]
                count = 1
                
        if count == 1:
            count = ""
        result += str(count) + tempStr
        length.append(len(result))
        result = ""
        
    return min(length)
```

```python
str1 = "aabbaccc" # 7
print(solution2(str1))
```

```python
str2 = "ababcdcdababcdcd" # 9
print(solution2(str2))
```

```python
str3 = "abcabcdede" # 8
print(solution2(str3))
```

```python
str4 = "abcabcabcabcdededededede" # 14
print(solution2(str4))
```
